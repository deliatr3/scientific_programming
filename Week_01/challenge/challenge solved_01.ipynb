```python
import numpy as np #import numpy library
import matplotlib.pyplot as plt #import matplotlib library
array = np.array([1,2,3,4,5,6,55,100,333]) #create an array

q1 = np.percentile(array, 25) #25th percentil #calculate the percentiles
q2 = np.percentile(array, 50) #50th percentil #calculate the percentiles
q3 = np.percentile(array, 75) #75th percentil #calculate the percentiles

print(q1) #print the percentiles    
print(q2)#print the percentiles 
print(q3)  #print the percentiles   


plt.hist(array, bins=10, color='c', edgecolor='black') #plot the histogram
plt.title("Histogram of array") #title of the histogram
plt.show() #display the histogram 

plt.boxplot(array) #plot the boxplot
plt.title("Boxplot of array") #title of the boxplot
plt.show() #display the boxplot 



```
